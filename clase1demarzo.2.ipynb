```python
df = pd.read_excel('datasets/Canada.xlsx',
                   engine = 'openpyxl',
                   sheet_name = 'Canada by Citizenship',
                   skiprows = range(20),
                   skipfooter = 2
                  )
df.head()
```

```python
df.drop(['Type', 'Coverage', 'AREA', 'REG', 'DEV', 'DevName'],
        axis =1,
        inplace = True
       )
```

```python
df.columns
df.rename(columns = {'OdName': 'Country', 'AreaName': 'Continent', 'RegName': 'Region'},
         inplace = True
         )
```

```python
df.columns = list(map(str, df.columns))
df.columns
```

```python
df.set_index('Country', inplace = True)
df.head()
            
```

```python
df['Total'] = df.iloc[:,
                      3:].sum(axis=1)
df.head(2)
```

```python
df.iloc[:,2]
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#pip installmatplotlib
mpl.style.use('ggplot')
```

```python
print(plt.style.available)
```

```python
j_data = df.loc['Japan', '1980' : '2013']
j_data.plot()
```

```python
ic_data = df.loc[['India', 'China'], '1980' : '2013']
nic = ic_data.transpose()
```

```python
nic.plot(figsize = (20,10))
plt.xlabel('Years', fontsize =17)
plt.xticks(fontsize =17, rotation=45)
plt.ylabel('Numer of inmigrats', fontsize =17)
plt.yticks(fontsize =17)
plt.title('Inmigration to Canada from India and China during 1980 to 2013', fontsize =17)
```
