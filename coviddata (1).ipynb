# Description of data set

The dataset that is being analyzed, was obtained form the CDMX state goverment. 

# My variables of interest

fecha_toma_muestra : Es el dato que refiere a la fecha de toma de muestra 
pruebas_totales: Total de pruebas en la Ciudad de México en el día
positivos_totales: Total de personas en la Ciudad de México con prueba positiva a SARS-CoV-2 en el día 
pruebas_totales_cdmx: Total de pruebas de residentes de la Ciudad de México en el día
positivos_totales_cdmx: Total de personas residentes en la Ciudad de México con prueba positiva a SARS-CoV-2 en el día 
tasa_positividad: Tasa de positividad diaria de personas en Ciudad de México 
tasa_positividad_cdmx: Tasa de positividad diaria de personas residentes en la Ciudad de México
