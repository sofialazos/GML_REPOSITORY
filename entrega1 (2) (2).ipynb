### Description of dataset
The following database shows an annual record of the population in New York, the water consumption and the consumption per capita

```python
df = pd.read_excel('Water_Consumption_in_the_City_of_New_York.xlsx',
                   engine = 'openpyxl',
                   #sheet_name ='casos_positivos',
                   #skiprows = range(20),
                   #skipfooter = 2
                  )
df.head(5)
```

### Variables of interest 
1. Year : cuantitative, discrete, average data registered per year
2. New York City Population: cuantitative, discrete population in NY 
3. NYC Consumption(Million gallons per day):  cuantitative, continue Water consumption in million per gallons per day in NYC 
4. Per Capita(Gallons per person per day): cuantitative, discrete Consumption per capita 

#### How is the behavior of the data between the consumption of water through the years?

```python
df3 = df[['Year','NYC Consumption(Million gallons per day)']]
```

```python
df3.plot(x='Year',y='NYC Consumption(Million gallons per day)',kind='line')
```

Through the graph it can be seen that water consumption decreases as the years go by

#### What are the ranges of water consumption in NYC?

```python
sns.boxplot(x = df['NYC Consumption(Million gallons per day)'], 
            whis=(0, 100)).set_title("NYC Consumption")

plt.show()
```

Through the box plot, it is observed that the maximum amount of water consumed is around 1,500 gallons and the minimum is below 1,000 gallons. Also, the average of water consumption is around 1200 gallons.

#### What is the trend in water consumption as the population in NYC increases?

```python

# Create Scatterplot
sns.lmplot(x='New York City Population', 
           y='NYC Consumption(Million gallons per day)', 
           data=df
          )
plt.title(' NYC Population vs NYC Consumption')
plt.show()
```

Water consumption in NYC tends to decrease as the population increases


#### How is the behavior of water consumption per capita in NYC?

```python
# Plot a histogram of the total bill
sns.histplot(df["Per Capita(Gallons per person per day)"], 
             kde = True).set_title("Histogram of Per capita consumption")
plt.show()
```

It can be observed that most people consume around 120 gallons of water per day. On the other hand, the minority consumes 160 gallons of water. Also,  it can be seen that it follows a senoidal trend because the frecuency increases and decreases as the consumtion of water per capita inreases
