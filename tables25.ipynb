```python
import seaborn as sns
```

```python
%ls datasets
```

```python
df=pd.read_csv('datasets/''Cartwheeldata.csv')
df.head()
```

```python
sns.lmplot(x='Wingspan',
           y='CWDistance',
           data=df,
           fit_reg=False,
           hue='Gender'
          )
plt.title("Scatter plot of CWDistance vs Wingspan")
```

```python
sns.swarmplot(x='Gender',
             y='CWDistance',
             data=df
            )
```

```python
sns.boxplot(data=df.loc[:,['Age','Height','Wingspan','CWDistance','Score']])
plt.xticks(rotation=45, fontsize=17)
plt.yticks(fontsize=17)
plt.title('Box Plot representation of the entire data', fontsize=15)
plt.show()
```

```python
#box plot of male data
```

```python
sns.boxplot(data=df.loc[df['Gender']=='M',
                        ['Age','Height','Wingspan','CWDistance','Score']])
plt.xticks(rotation=45, fontsize=17)
plt.yticks(fontsize=17)
plt.title('Box Plot representation of the entire data', fontsize=15)
plt.show()
```

```python
fig=plt.figure()
ax1=fig.add_subplot(1,2,1)#(#of rows ,#of columns, position)

sns.boxplot(data=df.loc[df['Gender']=='M',
                        ['Score']
                       ]
           )
ax2=fig.add_subplot(1,2,2)#(#of rows ,#of columns, position)

sns.boxplot(data=df.loc[df['Gender']=='F',
                        ['Score']
                       ]
           )

ax1.title.set_text('Male box plot')
ax2.title.set_text('Female box plot')
plt.show()
```

```python
sns.distplot(df.CWDistance)
plt.show()
```

```python
sns.countplot(x='Gender', data=df)
plt.show()
```
